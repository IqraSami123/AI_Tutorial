```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('retail_sales.csv ')
print(df.head())
```

```python
# sort by quantity in ascending, then by price in descending order
sorted_df = df.sort_values(by=['Quantity', 'Price'], ascending=[True, False])
print(sorted_df.head())
```

```python
# ranking products by quantity sold
df['Quantity_Rank'] = df['Quantity'].rank(method='min', ascending=False)
print(df[['Product', 'Quantity', 'Quantity_Rank']].head())
```
