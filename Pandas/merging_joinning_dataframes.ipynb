```python
import pandas as pd
```

```python
# samples dataframes
df1 = pd.DataFrame({
    'userID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})
df2 = pd.DataFrame({
    'userID': [2, 3, 3],
    'books': ['ENG', 'ISLAMIYAT', 'MATH']
})
```

```python
# inner join
inner_join = pd.merge(df1, df2, on='userID', how='inner')
print("Inner Join:\n", inner_join)
```

```python
# left join
left_join = pd.merge(df1, df2, on='userID', how='left')
print("\nLeft Join:\n", left_join)
```

```python
# concatinate rows example
df3 = pd.DataFrame({
    'userID': [4, 5],
    'Name': ['David', 'Eve']
})
concat_rows = pd.concat([df1, df3], ignore_index=True)
print("\nConcatenated Rows:\n", concat_rows)

# for concatenation it is important to have same columns in both dataframes otherwise it will fill missing values with NaN
```
