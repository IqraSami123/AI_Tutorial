```python
import pandas as pd
import openpyxl

df = pd.read_csv('retail_sales.csv', parse_dates=['Date'],
dtype={'Store': 'category', 'Product': 'category', 'Revenue': 'float'})
df.info()
```

```python
clean_df = df[df['Quantity'] > 1]  # filter rows where Quantity is greater than 1
subset = clean_df[['Category', 'Quantity', 'City']]  # select only the Category, Quantity, and City columns
# remove invalid rows where Quantity = 0 or negative

subset_to_excel = subset.to_excel('cleaned_retail_sales.xlsx', index=False, sheet_name='Cleaned Data'  )  # save the cleaned subset to an Excel file without the index
subset_to_json = subset.to_json('cleaned_retail_sales.json', orient='records', date_format='iso' )  # save the cleaned subset to a JSON file with records orientation

print("Cleaned data saved to Excel and JSON files successfully.")
print(subset.info())
```
