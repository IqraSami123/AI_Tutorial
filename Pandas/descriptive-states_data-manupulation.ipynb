```python
import pandas as pd
```

```python
df = pd.read_csv('ecommerce_sales_analytics_5000.csv', parse_dates=['order_date'])
print(df.head())
```

```python
# verify dataset before analysis
df.shape
```

```python
df.describe()
```

```python
df[['unit_price', 'revenue']].describe()    
```

```python
# quantiles
df['revenue'].quantile([0.25, 0.5, 0.75])
```

```python
# means and median 
print("Mean Revenue:", df['revenue'].mean())
print("Median Revenue:", df['revenue'].median())
# indetifying skewness
df['revenue'].skew()      # positive skewness indicates a long tail on the right side of the distribution, while negative skewness indicates a long tail on the left side. A skewness value close to 0 suggests a relatively symmetric distribution.
```

```python
# category level descriptive statistics
df.groupby('product_category')['revenue'].describe()
```

```python
# find highest mean 
df.groupby('product_category')['revenue'].mean()
```

```python
# using IQR to understand spread and identify outliers
Q1 = df['revenue'].quantile(0.25)
Q3 = df['revenue'].quantile(0.75)
IQR = Q3 - Q1
Q1, Q3, IQR

```
