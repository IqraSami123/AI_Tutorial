```python
import pandas as pd
import openpyxl

df = pd.read_csv('retail_sales.csv', parse_dates=['Date'],
dtype={'Category': 'category'})
df.info()
```

```python
# convert product datatype to category for memory efficiency
df['Product'] = df['Product'].astype('category')
df.info()
print(df['Product'].memory_usage(deep=True))    # memory usage of the Product column after conversion to category
```
