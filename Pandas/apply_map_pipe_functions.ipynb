```python
import pandas as pd
```

```python
df = pd.read_csv('ecommerce_sales_analytics_5000.csv', parse_dates=['order_date'])
print(df.head())
```

```python
# varify
print(df.info())
```

```python
# map() label high value orders
df['payment_type'] = df['payment_method'].map({
    'Card': 'Digital',
    'Wallet': 'Digital',
    'COD': 'Cash'
})
print(df[['payment_method', 'payment_type']].head())
# Explanation :   each value is mapped to a new value based on the provided dictionary. For example, 'Card' and 'Wallet' are mapped to 'Digital', while 'COD' is mapped to 'Cash'.
```

```python
# verify
df['payment_type'].unique()
```

```python
df.groupby('payment_type').size()
```

```python
# apply a custom function to create a reveneue category based 
def revenue_category(value):
    if value > 500:
        return 'High'
    elif value >= 200:
        return 'Medium'
    else:
        return 'Low'
df['revenue_category'] = df['revenue'].apply(revenue_category)
print(df['revenue_category'].value_counts())

# why apply() ?  depends on columns   as map() is used for single column mapping while apply() can be used for more complex operations that may involve multiple columns or custom logic.
```

```python
df[['revenue', 'revenue_category']].sample(10)
```

```python
# pip()   Metod chaining
# allows you to chain multiple operations together in a more readable and concise way. It is often
def add_profit(data):
    data['profit'] = data['revenue'] * 0.20
    return data
def net_revenue(data):
    data['net_revenue'] = data['revenue'] - data['profit']
    return data


df =(
    df
    .pipe(add_profit)
    .pipe(net_revenue)
)

# explanation: pipe allows you to apply a function to the DataFrame in a chainable manner. Each function takes the DataFrame as input, performs its operation, and returns the modified DataFrame, which can then be passed to the next function in the chain.
```
