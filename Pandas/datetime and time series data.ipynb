```python
import pandas as pd
```

```python
df = pd.read_csv('ecommerce_sales_analytics_5000.csv', parse_dates=['order_date'])
print(df.head())
```

```python
# setting date time indexing
df.set_index('order_date', inplace=True)
print(df.head())
```

```python
df.index
```

```python
# monthly revenue analysis 
monthly_revenue = df['revenue'].resample('ME').sum()
print(monthly_revenue)
```

```python
# trend interpretation
monthly_revenue.pct_change()
```

```python
# verify that index is monotonically increasing
monthly_revenue.index.is_monotonic_increasing
```

```python
# check that monthly reveue is greather than total revenue
print(monthly_revenue.sum() , df['revenue'].sum())
monthly_revenue.sum()  <= df['revenue'].sum()
```
