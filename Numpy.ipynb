```python
import numpy as np  # this creates a 2D array from a list of lists

arr = np.array([[2, 3, 4, 5], [5, 7, 8, 9]])
print(arr)

print(arr[-1, -2])  # prints the value at last row and second-to-last column
print(arr[1:2, 2:3])  # slicing on the 2 domentional array 
print(arr[::2, ::2])   # print every second row and every second colimn 
```

```python
import numpy as np  # this creates a 2D array from a list of lists

arr = np.array([[2, 3, 4, 5], [5, 7, 8, 9]])
print(arr)

new_array = arr[0:1, 1:3]
print(arr[0,1])
new_array[0, 1] = 100     # this will change the value at 0th row and 2nd column of new_array to 100 in original array
print(new_array) 
print(arr)
 # the original array is also modified because new_array is a view of arr
```

```python
import numpy as np  # this creates a 2D array from a list of lists

arr = np.array([[2, 3, 4, 5], [5, 7, 8, 9]])
print(arr)

new_array = arr[0:1, 1:3].copy() # this will create a copy of the sliced array, so changes to new_array will not affect arr
print(arr[0,1])
new_array[0, 1] = 100    
print(new_array) 
print(arr)
```

```python
rang = np.random.default_rng()  
flips  =  rang.integers(0, 2, size=100000)  # generate 10 random integers between 0 and 1 (inclusive)
prob_head = np.mean(flips)
print("probality heads are:", prob_head)
```

```python
rang = np.random.default_rng(seed=5)  # setting a seed for reproducibility for exact same random numbers every time we run the code
flips  =  rang.integers(0, 2, size=100000)  # generate 10 random integers between 0 and 1 (inclusive)
prob_head = np.mean(flips)
print("probality heads are:", prob_head)
```
