```python
import pandas as pd
```

```python
import numpy as np
df = pd.read_csv("HousePricePrediction.csv")
print(df.head(10))
```

```python
print("Shape of Dataset:", df.shape)
print("\nData Types:\n")
print(df.dtypes)
print("\nSummary Statistics:\n")
print(df.describe())
```

```python
# Check missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# Fill missing numerical values with mean
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill missing categorical values with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Check duplicates
print("\nDuplicate Rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()
```

```python
# Remove unnecessary columns such as Id (if exists)
if 'Id' in df.columns:
    df = df.drop('Id', axis=1)

# Display remaining columns
print("\nSelected Features:\n")
print(df.columns)
```

```python
# Convert categorical variables into numerical form
df_encoded = pd.get_dummies(df, drop_first=True)

# Display preview of processed dataset
print("\nProcessed Dataset Preview:\n")
print(df_encoded.head())

# Display final shape
print("\nFinal Dataset Shape:", df_encoded.shape)
```

```python
# Save cleaned and encoded dataset
df_encoded.to_csv("Processed_HousePricePrediction.csv", index=False)

print("Processed dataset saved successfully.")
```
