```python
import pandas as pd
s = pd.Series([12,30,40,50,23, 30],index = [1,2,3,4,5, 6]) 
print("indexes of the seriese are:", s.index) # this prints the indexes of the series
print("values of the seriese are:", s.values) # this prints the values of the series
print("datatype of the seriese are:", s.dtype) # this prints the datatype of the series
print("head of the series is:", s.head(2)) # this prints the first 5 elements of the series
print("tail of the series is:", s.tail(3)) # this prints the last 5 elements of the series
print("count of values in the series is:", s.value_counts()) # this counts the number of occurrences of each unique value in the series
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 35, 40, 45],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
        'salary': [50000, 60000, 70000, 80000, 90000]}
df = pd.DataFrame(data) # this creates a pandas dataframe from the given dictionary
print(df)
```

```python
#label based selection  (.loc)  (both forst and last index are included)
print(df.loc[0]) # this selects the first row of the dataframe using label based indexing
print(df.loc[0:2]) # this selects the first three rows of the dataframe using label based indexing
print(df.loc[:, 'Name']) # this selects the 'Name' column of the dataframe using label based indexing
print(df.loc[0:2, ['Name', 'Age']]) # this selects the first three rows and 'Name' and 'Age' columns of the dataframe using label based indexing
```

```python
# integer based selection  (.iloc) (both first and last index are not included, only the first index is included)
print(df.iloc[0]) # this selects the first row of the dataframe using integer based indexing
print(df.iloc[0:2]) # this selects the first three rows of the dataframe using integer based indexing
print(df.iloc[:, 0]) # this selects the first column of the dataframe using integer based indexing
print(df.iloc[0:2, [0, 1]]) # this selects the first three rows and first two columns of the dataframe using integer based indexing         
```

```python
# boolean indexing
print(df[df['Age'] > 30]) # this selects the rows of the dataframe where the value of the 'Age' column is greater than 30
print(df[df['City'] == 'New York']) # this selects the rows of the dataframe where the value of the 'City' column is 'New York' 
```

```python
# safe assignment using .loc
df.loc[df['Age'] > 30, 'Senior'] = True # this creates a new column 'Senior' and assigns the value True to the rows where the value of the 'Age' column is greater than 30
print(df)       
df.loc[df['Age'] > 30, 'Salary']= 2000000000 # this updates the value of the 'Salary' column to 2000000000 for the rows where the value of the 'Age' column is greater than 30  
print(df)
```
